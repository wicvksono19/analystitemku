```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Caller and Agent Data.xls')
```

```python
df.dtypes
```

```python
df['Date'].ffill(inplace=True)
```

```python
df.set_index('Date', inplace= True)
```

```python
df
```

```python
df['CD'] = df.groupby('Date')['Completed Calls'].cumsum()
```

```python
plt.figure(figsize=(16,10))
plt.bar(df.index.to_list(),df['CD'])
plt.xticks(df.index.to_list())
plt.xticks(rotation=45)
plt.show()
```

```python
df.rename(columns={'Time (Local)':'Time'}, inplace=True)
df
```

```python
df.rename(columns={'Requests ':'Requests'}, inplace=True)
df
```

```python
df['TR'] = df.groupby('Time')['Requests'].cumsum()
```

```python
plt.figure(figsize=(16,10))
plt.bar(df['Time'],df['TR'])
plt.xticks(df['Time'])
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns
```

```python
df.rename(columns={'Unique Agents':'Unique'}, inplace=True)
df
```

```python
df['TU'] = df.groupby('Time')['Unique'].cumsum()
```

```python
plt.figure(figsize=(16,10))
plt.bar(df['Time'],df['TU'])
plt.xticks(df['Time'])
plt.xticks(rotation=45)
plt.show()
```

```python
df.rename(columns={'Zeroes ':'Zeroes'}, inplace=True)
```

```python
df['DZ'] = df.groupby('Date')['Zeroes'].cumsum()
```

```python
plt.figure(figsize=(16,10))
plt.bar(df.index.to_list(),df['DZ'])
plt.xticks(df.index.to_list())
plt.xticks(rotation=45)
plt.show()
```

```python
df.rename(columns={'Eyeballs ':'Eyeballs'}, inplace=True)
```

```python
df['TE'] = df.groupby('Time')['Eyeballs'].cumsum()
```

```python
plt.figure(figsize=(16,10))
plt.bar(df.Time,df['Eyeballs'])
plt.xticks(df.Time)
plt.xticks(rotation=45)
plt.show()
```
